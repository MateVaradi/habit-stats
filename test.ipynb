```python
# Imports
import numpy as np
import july
from july.utils import date_range
import seaborn as sns
import pandas as pd
import datetime

from habitstats.main import process_data
from habitstats.plots import heatmap_per_habit, heatmap_all_habits, barchart_all_habits
from habitstats.aesthetics import generate_color_dict
```

```python
# Create fake raw data
n = 500
all_dates = date_range("2022-01-01", "2022-12-31")
dates = sorted(np.random.choice(all_dates, n))
habits = np.random.choice(["s","g","x","m","a"], n)
data = pd.DataFrame({"date":dates,
                     "habit":habits})
data = process_data(data)
```

```python
habits = data.habit.unique()
color_dict = generate_color_dict(habits)
```

```python
heatmap_per_habit(data, 
                  "s", 
                  color=color_dict["s"],
                  title="S")
```

```python
heatmap_all_habits(data, by="week", 
                   colors=color_dict,
                   labels=True,
                   title="Habits per week")
```

```python
barchart_all_habits(data, "month", 
                    color_dict=color_dict,
                    title="Monthly frequency per habit",
                    )
```
